```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd

# 1. Setup Chrome browser
options = Options()
options.add_argument("--start-maximized")
options.add_argument("user-data-dir=selenium")  # Keep cookies between runs (make sure you're logged in)

service = Service('/path/to/chromedriver')  # üîÅ Replace with the path to your chromedriver
driver = webdriver.Chrome(service=service, options=options)

# 2. List of Instagram post URLs
post_urls = [
    "https://www.instagram.com/p/DKWeH34vxor/",
    "https://www.instagram.com/p/DIaBmQ3MOhl/",
    "https://www.instagram.com/reel/DIA339-MBUv/",
    # ... add up to 30 URLs here
]

# 3. Define post data container
data = []

# 4. Scraping loop
for url in post_urls:
    driver.get(url)
    time.sleep(5)  # Wait for content to load

    try:
        caption_elem = driver.find_element(By.XPATH, '//div[@data-testid="post-comment-root"]/div/div/span')
        caption = caption_elem.text
    except:
        caption = "No caption"

    try:
        likes_elem = driver.find_element(By.XPATH, '//section//button/span')
        likes = likes_elem.text.replace(",", "")
    except:
        likes = ""

    try:
        views_elem = driver.find_element(By.XPATH, '//span[contains(text(),"views")]')
        views = views_elem.text.split()[0].replace(",", "")
    except:
        views = ""

    try:
        comments_elem = driver.find_elements(By.XPATH, '//ul/ul')
        comments_count = len(comments_elem)
    except:
        comments_count = 0

    post_format = "Reel" if "/reel/" in url else "Image"

    data.append({
        "URL": url,
        "Caption": caption,
        "Format": post_format,
        "Views": views,
        "Likes": likes,
        "Comments": comments_count
    })

# 5. Save to CSV
df = pd.DataFrame(data)
df.to_csv("instagram_post_data.csv", index=False)
driver.quit()
```
